### Exercise 1

What is the plug-in principle and how is it used in non-parametric statistics?

The plug-in principle is where we approximate the distribution function of the generative model with the empirical distribution function from our data. We can estimate parameter values by finding the arithmetic mean of the statistic of interest from our data and plugging that in as our parameter.

### Exercise 2

What is a bootstrap sample and what is a bootstrap replicate?

A bootstrap sample is a set of n values drawn from your data with replacement. A bootstrap replicate is the calculated parameter value/plug-in estimate from the bootstrap sample.

### Exercise 3

Consider the following data set for waiting times in minutes for nuclear localization events of the gene MSN-2 in yeast (these are real measurements done by Yihan Lin in Michael Elowitz's lab).

```python
t = [3, 6, 11, 5, 5, 4, 73, 31, 7, 6, 
     30, 4, 32, 30, 5, 53, 2, 15, 18, 
     14, 3, 49, 7, 4, 4, 2, 9, 11, 8, 
     5, 14, 6, 32, 40, 3, 5, 24]
```

You suspect that MSN-2 localization may be a Poisson process and therefore that these waiting times are Exponentially distributed with a characteristic wait time of 20 minutes. Do a quick graphical analysis to check out this idea.

```python
t_gen = np.random.exponential(20, len(t))

p = bebi103.viz.ecdf(t, x_axis_label='t (min)', formal=True, line_width=2)
p = bebi103.viz.ecdf(t_gen, p=p, color='gray', formal=True, line_width=2)

bokeh.io.show(p)
```

```python
p = bebi103.viz.ecdf(t_gen, x_axis_label='t (min)', color='gray', alpha=0.1, formal=True)
 
for _ in range(100):
    t_gen = np.random.exponential(20, len(t))
    p = bebi103.viz.ecdf(t_gen, p=p, color='gray', alpha=0.1, formal=True)
    
p = bebi103.viz.ecdf(t, p=p, formal=True, line_width=2)

bokeh.io.show(p)
```

It looks like the waiting times are exponentially distributed.
