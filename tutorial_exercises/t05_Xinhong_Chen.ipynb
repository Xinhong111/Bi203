### Exercise 1

What is the plug-in principle and how is it used in non-parametric statistics?



Plug-in principle is the method of estimation of functionals of a population distribution by evaluating the same functionals at the empirical distribution based on a sample, it offers you the chance to estimate the parametr of whole population based on some collected samples. In non-parametric statistic, since there are no pre-assumptions of the distribution for the whole dataset, by using plug-in principle, we could get a non-bias idea of the real distribution of the whole dataset.

### Exercise 2

What is a bootstrap sample and what is a bootstrap replicate?

Bootstrap sample is a resampling process where large numbers of smaller samples of the same size are repeatedly drawn, with replacement, from a single original sample. Bootstrap replicate is one of the samller samples.

We have 37 measurements, and, as we have been doing, we will assume that the data are generated from a Exponentialdistribution. We can then simulate the data generation process by drawing 37 samples out of a Gaussian distribution. In so doing, we can see how we might expect the results from the experiment to vary if we were to do it again. (modified from tutorial)

```python
t_gen = np.random.exponential(scale = 20, size = len(t))
```

```python
p = bebi103.viz.ecdf(t, x_axis_label='t', formal=True, line_width=2)
p = bebi103.viz.ecdf(t_gen, p=p, color='gray', formal=True, line_width=2)

bokeh.io.show(p)
```

```python
beta = 20

# Make the draws and make the plot
t_gen = np.random.exponential(scale=beta, size=len(t))

p = bebi103.viz.ecdf(t_gen, x_axis_label='t', color='gray', alpha=0.1, formal=True)
 
for _ in range(100):
    t_gen = np.random.exponential(scale=beta, size=len(t))
    p = bebi103.viz.ecdf(t_gen, p=p, color='gray', alpha=0.1, formal=True)
    
p = bebi103.viz.ecdf(t, p=p, formal=True, line_width=2)

bokeh.io.show(p)
```
