**EX1**

The plug-in principle is that we can estimate statistical functionals by computing the functionals at the empirical distribution, which means we can plug in the data acquired and suppose it totally represents the distribution based on which we compute the functionals. In non-parametric statistics where we have no idea about the model, we can use this principle to compute some representative functionals like mean, median by only using the data we have and form a statistical sense of the data distribution.

**EX2**

A bootstrap sample is a set of sample values drawn the the original dataset with replacement, i.e. one sample value in the original dataset could be drawn multiple times into the bootstrap sample. A bootstrap replicate is the estimate of speicific functionals from the bootstrap sample.
